```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import csv
from sklearn import svm
from sklearn.neural_network import MLPClassifier
```

```python
pd_train = pd.read_csv('train.csv')
pd_test = pd.read_csv('test.csv')

pd_train = pd_train.drop("id",axis=1)
pd_test_ids = pd_test.loc[:,"id"]
pd_test = pd_test.drop('id',axis=1)

ytrain = pd_train.loc[:,"type"]
pd_train = pd_train.drop('type',axis=1)


pd_train =pd_train.replace("blood",1/6)
pd_train =pd_train.replace("white",2/6)
pd_train =pd_train.replace("black",3/6)
pd_train =pd_train.replace("green",4/6)
pd_train =pd_train.replace("blue",5/6)
pd_train = pd_train.replace(to_replace="clear",value=6/6)


pd_test =pd_test.replace("blood",1/6)
pd_test =pd_test.replace("white",2/6)
pd_test=pd_test.replace("black",3/6)
pd_test =pd_test.replace("green",4/6)
pd_test =pd_test.replace("blue",5/6)
pd_test = pd_test.replace(to_replace="clear",value=6/6)



Xtrain = pd_train.iloc[:,:].values


xtest = pd_test.iloc[:,:].values
```

```python
def printer(nameoffile,data):
    header = ['id', 'type']



    with open(nameoffile, 'w', encoding='UTF8', newline='') as f:
        writer = csv.writer(f)

    # write the header
        writer.writerow(header)
        for i in range(len(data)):
    # write the data
            writer.writerow(data[i])
```

# Naive Bayes Classifier

```python
pd_train = pd.read_csv('train.csv')
pd_train = pd_train.drop("id",axis=1)
pd_train =pd_train.replace("blood",1/6)
pd_train =pd_train.replace("white",2/6)
pd_train =pd_train.replace("black",3/6)
pd_train =pd_train.replace("green",4/6)
pd_train =pd_train.replace("blue",5/6)
pd_train = pd_train.replace(to_replace="clear",value=6/6)
# For all dataset
bone_mean = pd_train["bone_length"].mean()
rotting_flesh_mean = pd_train["rotting_flesh"].mean()
hair_length_mean = pd_train["hair_length"].mean()
has_soul_mean = pd_train["has_soul"].mean()
color_mean = pd_train["color"].mean()

bone_std = pd_train["bone_length"].std()
rotting_flesh_std = pd_train["rotting_flesh"].std()
hair_length_std = pd_train["hair_length"].std()
has_soul_std = pd_train["has_soul"].std()
color_std = pd_train["color"].std()

# For each class
ghoul_mean = pd_train.loc[pd_train["type"]=="Ghoul"].mean()
goblin_mean = pd_train.loc[pd_train["type"]=='Goblin'].mean()
ghost_mean = pd_train.loc[pd_train["type"]=='Ghost'].mean()

ghoul_std = pd_train.loc[pd_train["type"]=="Ghoul"].std()
goblin_std = pd_train.loc[pd_train["type"]=='Goblin'].std()
ghost_std = pd_train.loc[pd_train["type"]=='Ghost'].std()

ghoul_size = len(pd_train.loc[pd_train["type"]=="Ghoul"])
goblin_size = len(pd_train.loc[pd_train["type"]=='Goblin'])
ghost_size = len(pd_train.loc[pd_train["type"]=='Ghost'])
print(ghoul_size, goblin_size, ghost_size)

classes = dict()

classes['Ghoul'] = [[ghoul_mean.bone_length,ghoul_std.bone_length],[ghoul_mean.rotting_flesh,ghoul_std.rotting_flesh],[ghoul_mean.hair_length,ghoul_std.hair_length],[ghoul_mean.has_soul,ghoul_std.has_soul],[ghoul_mean.color,ghoul_std.color],ghoul_size]
classes['Goblin'] = [[goblin_mean.bone_length,goblin_std.bone_length],[goblin_mean.rotting_flesh,goblin_std.rotting_flesh],[goblin_mean.hair_length,goblin_std.hair_length],[goblin_mean.has_soul,goblin_std.has_soul],[goblin_mean.color,goblin_std.color],goblin_size]
classes['Ghost'] = [[ghost_mean.bone_length,ghost_std.bone_length],[ghost_mean.rotting_flesh,ghost_std.rotting_flesh],[ghost_mean.hair_length,ghost_std.hair_length],[ghost_mean.has_soul,ghost_std.has_soul],[ghost_mean.color,ghost_std.color],ghost_size]
```

```python
from math import sqrt
from math import pi
from math import exp

def calculate_probability(x, mean, stdev):
	exponent = exp(-((x-mean)**2 / (2 * stdev**2 )))
	return (1 / (sqrt(2 * pi) * stdev)) * exponent


total_rows = ghoul_size + goblin_size + ghost_size
def determine_class(classes,row,total_rows):
    
    
    
    probabilities = dict()
    
    for class_value in classes:
        for item in range(len(classes[class_value])-1):
            probabilities[class_value] = classes[class_value][-1]/total_rows
            for i in range(len(classes[class_value])-1):
                mean,std = classes[class_value][i]
                probabilities[class_value] *= calculate_probability(row[i],mean,std)
    
    return probabilities
    
    
total_probs = []   
for item in xtest:  
    proba = determine_class(classes,item,total_rows)   
    #print(proba)
    total_probs.append(proba)

total = []
data=[]

for i in total_probs:
    total.append(max(i, key = i.get))
for i in range(len(total)):
    temp = [pd_test_ids[i],total[i]]
    data.append(temp)
printer('bayes.csv',data)
```
